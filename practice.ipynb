You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1. If a string is longer than the other, append the additional letters onto the end of the merged string.

Return the merged string.

Example 1:

Input: word1 = "abc", word2 = "pqr"
Output: "apbqcr"
Explanation: The merged string will be merged as so:
word1:  a   b   c
word2:    p   q   r
merged: a p b q c r

```python
word1, word2 = "abcd", "pqde"

# Get the number of characters in each word
l1 = len(word1)
l2 = len(word2)
lmax = max(l1,l2)

result = []

# Loop to the maximum lenght, checking if there is stil characthers available in each word
for i in range(lmax):
    if i < l1:
        result.append(word1[i])
    if i < l2:
        result.append(word2[i])

result = ''.join(result)
```

```python
import numpy as np

def viterbi_algorithm(characters, prior_probs, transition_probs):
    n = len(characters)  # Number of characters
    T = len(prior_probs)  # Number of time steps (length of sequence)

    # Initialize the dynamic programming (DP) table
    dp = np.zeros((n, T))  # dp[i][t] will store the maximum probability of ending at character i at time t
    backtrack = np.zeros((n, T), dtype=int)  # To store the backtrack path

    # Step 1: Initialization - Set the initial probabilities using the prior probabilities
    for i in range(n):
        dp[i][0] = prior_probs[i]

    # Step 2: Recursion - Fill the DP table
    for t in range(1, T):
        for j in range(n):  # For each character at time step t
            # Find the maximum probability of transitioning to character j from any character at time t-1
            max_prob, max_idx = max(
                (dp[i][t-1] * transition_probs[i][j], i) for i in range(n)
            )
            dp[j][t] = max_prob
            backtrack[j][t] = max_idx  # Keep track of the character with the max probability

    # Step 3: Backtracking - Find the optimal sequence by backtracking
    # Start from the character with the highest probability at the final time step
    optimal_seq = []
    last_char_idx = np.argmax(dp[:, T-1])
    optimal_seq.append(characters[last_char_idx])

    # Backtrack through the optimal path
    for t in range(T-1, 0, -1):
        last_char_idx = backtrack[last_char_idx][t]
        optimal_seq.append(characters[last_char_idx])

    # Reverse the sequence since we backtracked from the end
    optimal_seq.reverse()

    return optimal_seq


# Example Usage
characters = ['A', 'B', 'C']
prior_probs = [0.2, 0.4, 0.4]
transition_probs = [
    [0.5, 0.2, 0.3],
    [0.3, 0.5, 0.2],
    [0.4, 0.3, 0.3]
]

optimal_sequence = viterbi_algorithm(characters, prior_probs, transition_probs)
print("Optimal Sequence:", optimal_sequence)
```
